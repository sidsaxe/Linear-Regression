```python
#lets preprocess training_data and testing_data
#we first separate the independent variable and the dependent variable/label in both

x_train= training_data.drop('median_house_value', axis = 1)
y_train= training_data['median_house_value']

x_test= testing_data.drop('median_house_value',axis = 1)
y_test= testing_data['median_house_value']

```

```python
#import the Linear  regression model from scikit-learn
from sklearn.linear_model import LinearRegression

#initializing the model and naming it lin_reg_model
lin_reg_model= LinearRegression()


#training the model using the fit method that accepts x_train and y_train
lin_reg_model.fit(x_train,y_train)
```

```python
#lets use the trained model to predict the x_test data that it has not seen before
y_pred = lin_reg_model.predict(x_test)
print(y_pred)
```

```python
#evaluating our model using the mean squared error
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)

print(mse)
```

```python
#using R square metric

from sklearn.metrics import r2_score

r_square_score= r2_score(y_test,y_pred)

print(r_square_score)
```
